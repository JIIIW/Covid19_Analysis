```python
# 연도에 따른 중국인 관광객수 시각화
import matplotlib.pyplot as plt

# 중국 국적 데이터 필터링
condition = (df['국적'] == '중국')
df_filter = df[condition]

plt.plot(df_filter['기준년월'], df_filter['관광'])
plt.show()
```

* 히트맵 그래프 그리기

```python
# x축: 월, y축: 연, 그래프 내용: 관광객 숫자
df.head()
```

```python
# str.slice() 함수를 이용해 연도, 월 컬럼 만들기
df['년도'] = df['기준년월'].str.slice(0, 4)
df['월'] = df['기준년월'].str.slice(5, 7)
df.head()
```

```python
# 중국인 관광객 필터링
condition = (df['국적'] == '중국')
df_filter = df[condition]
df_filter.head()
```

```python
df_pivot = df_filter.pivot_table(values='관광',
                                 index='년도',
                                 columns='월')
df_pivot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))
sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
plt.title('중국 관광객 히트맵')
plt.show()
```

```python
'''
- Trend : 2010년 1월에서 2016년 5월로 오면서 색깔이 진해짐, 즉 2010년 1월 ~ 2016년 5월까지 관광객 수가 꾸준히 증가하는 추세
- Seasonality : 각 연도별 기준으로 볼 때 7, 8월이 대체로 진한 색깔, 즉 관광객이 가장 많이 방문하는 패턴을 보임, 그 다음은 4, 5, 9, 10월이 높은 패턴으로 나타남
- Event : 2015년 6~ 8월까지 관광객 수가 적음
- Event : 2017년 3~ 6월까지 관광객 수가 매우 적음
- Event : 2020년 2월부터 관광객 수가 급격히 줄어듦
'''
```

```python
'''
= 전체 정리
- 중국인 관광객적은 계절적으로 여름 > 봄, 가을 > 겨울 순으로 많이 방문
- 트렌드를 통해 중국인 관광객 숫자가 계속해서 증가
- 2015/6/1 ~ 2015/8/31 : 키워드(중국인관광객), 메르스 여파로 중국인 관광객 숫자가 줄어듦
- 2017/3/1 ~ 2017/3/31 : 키워드(중국인관광객), 사드 보복 조치 여파로 중국인 관광객 숫자가 줄어듦
- 2020/5/1 ~ 2020/5/31 : 키워드(중국인관광객), COVID19 여파로 중국인 관광객 숫자가 줄어듦, 외국인 관광객이 98% 급감

- 메르스의 영향은 일시적인 관광객 감소로 이어진 후 원래 패턴을 회복하는 형태로 보임
- 사드 보복 조치 영향은 이전의 패턴과 비교해서 그 여파가 지속적으로 영향을 미치는 것을 알 수 있음
'''
```

```python
# 반복문으로 상위 5개국에 대한 히트맵 그래프 그리기
import matplotlib.pyplot as plt
import seaborn as sns

cntry_list = ['중국', '대만', '홍콩', '일본', '미국']

for cntry in cntry_list:
    condition = (df['국적'] == cntry)
    df_filter = df[condition]
    
    df_pivot = df_filter.pivot_table(values='관광',
                                 index='년도',
                                 columns='월')
    
    plt.figure(figsize=(16, 10))
    sns.heatmap(df_pivot, annot = True, fmt = '.0f', cmap = 'rocket_r')
    plt.title('{} 관광객 히트맵'.format(cntry))
    
    plt.show()


```
